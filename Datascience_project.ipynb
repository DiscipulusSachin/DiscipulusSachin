```python
!pip install reportlab
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
```

```python
def create_sales_report(filename, sales_mean, sales_median, sales_var, sales_std):
    c = canvas.Canvas(filename, pagesize=letter)
    c.drawString(100, 750, 'Sales Report')
    c.drawString(100, 700, f'Mean Sales: {sales_mean:.2f}')
    c.drawString(100, 680, f'Median Sales: {sales_median:.2f}')
    c.drawString(100, 660, f'Variance Sales: {sales_var:.2f}')
    c.drawString(100, 640, f'Standard Deviation Sales: {sales_std:.2f}')
    c.showPage()
    c.save()

create_sales_report('sales_report.pdf', sales_mean, sales_median, sales_var, sales_std)
```
